**Arithmetic Operators**

```python
### 1. Write a program to perform addition, subtraction, multiplication, division, floor division, modulus, and exponentiation on two numbers.
a = 10
b = 5
print(f"Addition is: {a+b}")
print(f"Subtraction is: {a-b}")
print(f"multiplication is: {a*b}")
print(f"division is: {a/b}")
print(f"floor division is: {a//b}")
print(f"modulus is: {a%b}")
print(f"exponentiation is: {a**b}")
```

```python
### 2. Swap two numbers without using a third variable.
a = 2
b = 3
print(f"value of a before swapping is {a}")
print(f"value of b before swapping is {b}")
a = a + b
b = a - b
a = a - b
print(f"value of a after swapping is {a}")
print(f"value of b after swapping is {b}")
```

```python
### 3. Find whether a number is even or odd using the modulus operator.
num = int(input())
if num % 2 == 0:
  print(f"{num} is even number")
else:
  print(f"{num} is odd number")
```

```python
### 4. Reverse a three-digit number using arithmetic operators.
num = 3560
num1 = num
reverse = 0
while num > 0:
  digit = num % 10
  reverse = (reverse * 10) + digit
  num //= 10
print(f"reverse of {num1} is {reverse}")
```

```python
### 5. Find the last digit and first digit of a given number.
num = 2456
i = 1
while num > 0:
  digit = num % 10
  if i == 1:
    print(f"last digit is {digit}")
  num //= 10
  if num == 0:
    print(f"first digit is {digit}")
  i += 1



```

```python
### 21. Count how many bits are set to 1 in a number using bitwise operations.
num = 12
count = 0
while num > 0:
  if num & 1:
    count += 1
  num = num >> 1
print(count)
```

```python
### 27. Count how many vowels in a string belong to a predefined vowel list using `in`.
string = input("Enter a string: ")
vowels = ['a','e','i','o','u']
count = 0
for i in string:
  if i in vowels:
    count += 1
print(count)
```

```python
### 29. Create a grading system using comparison and logical operators.
marks = int(input("Enter the marks: "))
if marks >= 90 and marks <= 100:
  print("'A' Grade, Excellent!!!")
elif marks >= 80 and marks < 90:
  print("'B' Grade, Very Good!!!")
elif marks >= 70 and marks < 80:
  print("'C' Grade, Good!!!")
elif marks >= 60 and marks < 70:
  print("'D' Grade, Average!!!")
elif marks >= 50 and marks < 60:
  print("'E' Grade, Below Average!!!")
elif marks >= 0 and marks < 50:
  print("Fail!!!")
else:
  print("Invalid marks")
```
